```python
import sqlite3
predictionDatabase = sqlite3.connect('database.db')
cursor = predictionDatabase.cursor()
cursor.execute("PRAGMA foreign_keys = ON;")

```

```python
cursor.executescript("""
CREATE TABLE Race (
    raceId INTEGER PRIMARY KEY AUTOINCREMENT,
    year INTEGER,
    circuit TEXT
);

CREATE TABLE Driver (
    driverId INTEGER PRIMARY KEY AUTOINCREMENT,
    code TEXT UNIQUE NOT NULL,
    name TEXT,
    team TEXT,
    teamColour TEXT
);

CREATE TABLE Lap (
    lapId INTEGER PRIMARY KEY AUTOINCREMENT,
    raceId INTEGER,
    driverId INTEGER,
    lapNumber INTEGER,
    attacking INTEGER,
    defending INTEGER,
    clean INTEGER,
    FOREIGN KEY (raceId) REFERENCES Race(raceId)
    FOREIGN KEY (driverId) REFERENCES Driver(driverId)
);

CREATE TABLE Features (
    lapId INTEGER PRIMARY KEY,
    throttlePerc100 REAL,
    throttlePerc0 REAL,
    avCornerBrakeDistance REAL,
    throttleOscillation REAL,
    coastingPerc REAL,
    brakeChanges REAL,
    avSpeedCornerDiff REAL,
    avApexThrottle REAL,
    FOREIGN KEY (lapId) REFERENCES Lap(lapId)
);
""")

predictionDatabase.commit()
```

```python
cursor = predictionDatabase.cursor()
cursor.execute("SELECT * FROM 'Features' LIMIT 210,30")
tables = cursor.fetchall()
print(tables)
cursor.close()
```
